# Data Encoding - Converting categorical values into a numerical format


### One Hot Encoding(HOE) or Nominal Encoding

 each category is represented as a binary vector where each bit corresponds to a unique category.ie convert each category into a new binary column .

if you one feature has 1000 categories,then 1000 new binary feature(column) will be created.

Resulting dataset is a Sparse, b'coz most of the entries are Zeros (with only one 1/row for each category). so it leades to  overfitting issue

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import seaborn as sns
```

```python
#create a df
df=pd.DataFrame(
    {
        'color':['red','blue','green','green','red','blue']
    }
)
```

```python
df.head()
```

```python
#create an instance of ONE hot encoder

encoder=OneHotEncoder()

#perform fit and trandsform
encoder.fit_transform(df[['color']]) # fits the OneHotEncoder on the color column of the df DataFrame and then transforms it into a sparse matrix.
```

```python
#convert to an array
encoded_values=encoder.fit_transform(df[['color']]).toarray()
encoded_values
```

```python
encoded_df=pd.DataFrame(encoded_values,columns=encoder.get_feature_names_out())
encoded_df
```

The columns are named using encoder.get_feature_names_out(), which generates appropriate column names based on the original column names and the categories.

```python
pd.concat([df,encoded_df],axis=1)
```

The original DataFrame df is concatenated with encoded_df along the columns (axis=1), resulting in a new DataFrame that contains both the original data and the one-hot encoded columns

```python
#for new data 
encoder.transform([['blue']]).toarray()
```

transforms a new data point (in this case, [['blue']]) using the already-fitted encoder.

### convert tips datset to numerical format

```python
df2=sns.load_dataset('tips')
df2
```

we need to convert categorical features like sex,smoker,day,and time 

```python
tips_encodedValues=encoder.fit_transform(df2[['sex','smoker','day','time']]).toarray()
tips_encodedValues
```

```python
tips_encoded=pd.DataFrame(tips_encodedValues,columns=encoder.get_feature_names_out())
tips_encoded
```

```python
pd.concat([df2,tips_encoded],axis=1)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y =make_circles(n_samples=750,factor=0.3,noise=0.1)
```

make_circles: This function generates a large circle containing a smaller circle in 2D. 
Itâ€™s commonly used to create synthetic datasets for classification problems and to test clustering algorithms.
 especially those designed to handle non-linearly separable data.
factor=0.3: This controls the scale of the smaller circle relative to the larger circle. 
A factor of 0.3 means that the smaller circle will have 30% of the radius of the larger circle.
X: A 2D array of shape (n_samples, 2), where each row represents a point in the dataset with two features (x and y coordinates).
y: A 1D array of shape (n_samples,), containing the class labels (0 or 1) for each point. 
Points in the smaller circle will have one label, while those in the larger circle will have another.

```python
plt.scatter(X[:,0],X[:,1],c=y) #c parameter assigns different colors to points according to their class labels or values in y.
```

```python
#apply dbscan

dbscan=DBSCAN(eps=0.10)#eps parameter defines the maximum distance between two points for one to be considered as in the neighborhood of the other
dbscan.fit_predict(X)  #fits the  model to the data X and then predicts the cluster labels for each point in X.
dbscan.labels_
```

-1 is outlier

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
plt.scatter(X[dbscan.labels_ == -1, 0], X[dbscan.labels_== -1, 1], color='red', label='Outlier')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering of Circular Data')
plt.show()
```

outliers are marked as red color

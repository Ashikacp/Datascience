```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing inbuilt dataset from sklearn
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris.target
```

```python
#we can load iris dataset from seaborn aswell.choose either one
import seaborn as sns
df=sns.load_dataset('iris')
df
```

```python
#independent and dependent feature
X=df.iloc[:,:-1] # it will print independent feature by removing the output.ie species
y=iris.target # taking y from sklearn dataset
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
#post-pruning ie, grows the tree fully and then prune.
tree_model=DecisionTreeClassifier() #not giving any parameters. taking default parameter from the method. and it is post-pruning
tree_model.fit(X_train,y_train)
```

```python
#we got the model from the previous  step .now we are going to construct the DT
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_model,filled=True) #filled parameter helps to fill the DT boxes with colors
```

```python
#prediction
y_pred=tree_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

```python
#here for each class we are getting binary decision .it is called cart
#gini value is b/w 0 to 0.5 and  if gini=0 leaf node/pure node . o/p value setosa is leading (31). classified.
#  after x(2)<= 5.35 , we are unnecessary doing the split.we need to split theh DT upto X(2)<=2.5 and stop it.b'coz the value is [0,34,4].
#here the o/p value versicolor is a majority value(34) and there is a huge difference b/w versicolor and verginia ,34 and 4.
#so we need to add max_depth=2 in DecisionTreeClassifier()   ie,    tree_model=DecisionTreeClassifier(max_depth=2)

tree_model=DecisionTreeClassifier(max_depth=2) #not giving any parameters. taking default parameter from the method. and it is post-pruning
tree_model.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(tree_model,filled=True)
```

```python
#prediction
y_pred=tree_model.predict(X_test)
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

```python
#model accuracy will be changed after pruning . before 96 and after pruning accuracy became=98
```

```python
#pre-pruning
#add parameter in a dictionary.
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

tree_model=DecisionTreeClassifier()
cv=GridSearchCV(tree_model,parameter,cv=5,scoring='accuracy')#estimator= the model you want to trian, tree_model and param_grid=parameter
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(X_test)
score=accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

```python
import pandas as pd
df=pd.read_csv('healthcare.csv')
df.head()
```

this healthcare data has 2 features.here we try to create isolation forest.Multiple isolation trees.

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
clf=IsolationForest(contamination=0.2)  #Specifies that 20% of the data is expected to be anomalies or outliers. This helps the model set a threshold for deciding which points are anomalous.
#float range(0-0.5).above this threshold we consider the datapoint as an outlier
clf.fit(df)
predictions=clf.predict(df)
predictions
```

here fit(df) Trains the IsolationForest model using the data in df. The model learns patterns in the data and 
determines which points are more likely to be anomalies based on how easily they can be isolated.
After training, this method predicts whether each data point in df is an anomaly or not.

The predictions array contains 1 for normal data points and -1 for anomalies.

```python
import numpy as np
index_outlier=np.where(predictions<0)
index_outlier
```

```python
X=df.values
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1]) #Select the 1rst & 2nd features(columns) from the dataframe df for plotting.
plt.scatter(X[index_outlier,0],X[index_outlier,1],edgecolors='red',facecolors='none') #Plot outliers with red edges
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Isolation Forest: Anomaly Detection")
```

The plot displayed the main dataset with all points and highlight the outliers with red edges. This approach allows to visually inspect points which have been identified as anomalies by the IsolationForest algorithm.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
#load dataset

from sklearn.datasets import load_breast_cancer #inbuilt dataset of sklearn
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
cancer_dataset.keys()
```

```python
cancer_dataset.feature_names
```

```python
#create a dataframe

df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names']) # data key has the entire data 
df.head(5)
```

```python
'''here we have around 30 features and hence we can't visualise it.we need to convert this to 2D or 3D.
HERE we are  not going to remove the feature(feature selection) instead we try to extract the feature(feature extraction)
We will take all these features and try to extract only 2 feature that captures max amount of information out of it.
'''
```

```python
## standardisation
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(df)  #entire fit has happend on the entire dataset
```

```python
#perform transformation 
scaled_data=scalar.transform(df)  # standardised data with mean=0 and std.deviation=1
scaled_data
```

```python
#apply PCA algorithm

from sklearn.decomposition import PCA

pca=PCA(n_components=2)  #we need to extract only 2 featue out of 30 features ,which capture max variance

data_pca=pca.fit_transform(scaled_data)
data_pca
```

here we reduced the 30 features into 2 features(which has max variance ).ie, dimensionality reduction. The process of creating new feature from the original features is called feature extraction.Pca is a widely used technique for dimensionality reduction 

```python
pca.explained_variance_   #variance of new 2 features
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cancer_dataset['target'], cmap='plasma')
plt.xlabel("First Principal Component") 
plt.ylabel("Second Principal Component")  
plt.title("PCA of Cancer Dataset")
plt.show()
```

here data point is now clearly seperable since we were able to capture max amount of information

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.neighbors import LocalOutlierFactor
```

```python
X, y = make_circles(n_samples=300, noise=0.02, factor=0.3, random_state=42)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('make_circles dataset')
plt.show()
```

```python
#apply LOF
lof=LocalOutlierFactor(n_neighbors=20,contamination=0.1)
pred=lof.fit_predict(X)
pred
```

LocalOutlierFactor class is used to detect local outliers. 
Here, n_neighbors=20 specifies the number of neighbors to use for calculating the LOF score, and 
contamination=0.1 indicates that approximately 10% of the data is considered as outliers

pred provides labels (1 for inliers and -1 for outliers)

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:,0],X[:,1]) #or we can use plt.scatter(X[pred==1, 0],X[pred== 1, 1], color='blue', label='Inlier')
plt.scatter(X[pred==-1, 0],X[pred== -1, 1], color='red',edgecolor='black', label='Outlier')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Local Outlier Factor (LOF) on make_circles')
plt.legend()

plt.show()
```
